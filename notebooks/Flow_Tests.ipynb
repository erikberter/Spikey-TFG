```python
from PIL import Image
  
# open method used to open different extension image file
im_u = Image.open(path_u) 

rgb_im_u = im_u.convert('RGB') 
  
# This method will show image in any image viewer 
display(rgb_im_u) 
```

```python
from PIL import Image
  
# open method used to open different extension image file
im_v = Image.open(path_v) 

rgb_im_v = im_v.convert('RGB') 
  
# This method will show image in any image viewer 
display(rgb_im_v)
```

```python
# Split into 3 channels
r_u, _, _ = rgb_im_u.split()

_, g_v, b = rgb_im_u.split()

# Increase Reds
b = b.point(lambda i: 127)


# Recombine back to RGB image
result = Image.merge('RGB', (r_u, g_v, b))
display(result)
```

```python
resuilt_res = result.resize((172,128), Image.ANTIALIAS)
display(resuilt_res)
```
